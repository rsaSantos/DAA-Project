```python
# Import the required packages
# import warnings
# warnings.filterwarnings('ignore')

#Pandas stack
import pandas as pd
import pandas_profiling
#Plots
import matplotlib.pyplot as plt
import seaborn as sns
# Lois
from lois import lois_ds_report

%matplotlib inline
```

```python
# Load london airbnb data
data = pd.read_csv('../data/LondonAirbnb/LondonAirbnb.csv')
```

```python
# Automate the EDA with lois
lois_ds_report(data, target_variable='price')
```

```python
# Remove empty columns (license and neighbourhood_group)
data = data.drop(['license', 'neighbourhood_group'], axis=1)
```

```python
# Get simple pandas profiling report
data.profile_report(minimal=True)
```
